```python
#moving average for ap
df1['moving_avg'] = df1['ap'].rolling(window=20).mean()
df1.head(30)
```

```python
df1_s = pd.DataFrame()
```

```python
df1_s['tempr'] = np.mean(df1['tempr']) - df1['tempr']
df1_s['ap'] = np.mean(df1['ap']) - df1['ap']
df1_s['ws'] = np.mean(df1['ws']) - df1['ws']
df1_s['rh'] = np.mean(df1['rh']) - df1['rh']
df1_s['blizzard'] = np.mean(df1['blizzard'])- df1['blizzard']
df1_s.set_index(df1.index)
```

```python
sns.histplot(data=df1, x='tempr',color = 'navy')
sns.histplot(data=df1_s, x='tempr')
```
